```python
from IPython.display import display, Javascript, Audio
```

```python
from google.colab import output
import base64
```

```python
# JavaScript to record audio from the browser
def record_audio(seconds=5):
    js = Javascript('''
        async function recordAudio(seconds) {
            const stream = await navigator.mediaDevices.getUserMedia({ audio: true });
            const mediaRecorder = new MediaRecorder(stream);
            let chunks = [];
            mediaRecorder.ondataavailable = (event) => chunks.push(event.data);
            mediaRecorder.start();
            await new Promise((resolve) => setTimeout(resolve, seconds * 1000));
            mediaRecorder.stop();
            return new Promise((resolve) => {
                mediaRecorder.onstop = async () => {
                    const blob = new Blob(chunks);
                    const reader = new FileReader();
                    reader.readAsDataURL(blob);
                    reader.onloadend = () => resolve(reader.result);
                };
            });
        }
    ''')
    display(js)
    data = output.eval_js(f'recordAudio({seconds})')
    binary = base64.b64decode(data.split(',')[1])
    with open('command.wav', 'wb') as f:
        f.write(binary)
    return 'command.wav'
```

```python
# import libary and packages

import speech_recognition as sr
from gtts import gTTS
import os
import datetime
import webbrowser
```

```python

# 1. Text to Speech Function
def speak(text):
    print(f"Assistant: {text}")
    tts = gTTS(text=text, lang='en')
    tts.save("response.mp3")
    display(Audio("response.mp3", autoplay=True))

# 2. Command Processing Logic
def process_command(command):
    command = command.lower()

    if 'time' in command:
        now = datetime.datetime.now().strftime("%H:%M")
        speak(f"The current time is {now}")

    elif 'search' in command:
        search_query = command.replace("search", "").strip()
        url = f"https://www.google.com/search?q={search_query}"
        speak(f"Searching Google for {search_query}")
        print(f"Opening: {url}") # Note: Browser pop-ups might be blocked in Colab

    elif 'hello' in command or 'hi' in command:
        speak("Hello! I am your AI assistant. How can I help you today?")

    elif 'open youtube' in command:
        speak("Opening YouTube")
        print("Redirecting to: https://www.youtube.com")

    else:
        speak("I heard you, but I don't have a rule for that command yet.")

# 3. Main Execution Loop
def run_assistant():
    recognizer = sr.Recognizer()

    try:
        print("\n--- Listening... (Speak now) ---")
        audio_file = record_audio(seconds=4)

        with sr.AudioFile(audio_file) as source:
            audio_data = recognizer.record(source)
            text = recognizer.recognize_google(audio_data)
            print(f"User: {text}")
            process_command(text)

    except sr.UnknownValueError:
        speak("Sorry, I couldn't understand the audio.")
    except Exception as e:
        speak("An error occurred. Please check your microphone permissions.")
        print(f"Error: {e}")

# Display Sample Commands
print("""
Sample Commands you can try:
1. "What time is it?"
2. "Hello"
3. "Search Artificial Intelligence"
4. "Open YouTube"
""")
```

```python
run_assistant()
```
